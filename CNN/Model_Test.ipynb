```python
import cv2
import matplotlib.pyplot as plt
import os
import tensorflow as tf
import numpy as np

img_height,img_width=180,180

# Load the saved model
model_path = 'C:\\Users\\afiqe\\Downloads\\mamak'  # Replace with the actual path to your saved model
loaded_model = tf.keras.models.load_model(model_path)

# Define a dictionary to map class indices to class names
data_dir = 'C:\\Users\\afiqe\\Downloads\\DatasetNew'
class_names = sorted([d for d in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, d))])
print(class_names)

# Load and preprocess a single image for testing
test_image_path = 'C:\\Users\\afiqe\\Downloads\\roti canai.jpg'
img = cv2.imread(test_image_path)
image_resized = cv2.resize(img, (img_height, img_width))
image = np.expand_dims(image_resized, axis=0)
```

```python
# Make predictions
predictions = loaded_model.predict(image)

# Get the class with the highest probability
predicted_class_index = np.argmax(predictions)
predicted_class = class_names[predicted_class_index]

print("Predicted class:", predicted_class)

# Display the image with the predicted class label
plt.imshow(cv2.cvtColor(image_resized, cv2.COLOR_BGR2RGB))
plt.title("Predicted class: " + predicted_class)
plt.axis('off')
plt.show()
```
